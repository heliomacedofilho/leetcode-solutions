```python
class Solution:
    def shortestPathBinaryMatrix(self, grid: List[List[int]]) -> int:

        queue = []
        if grid[0][0] == 0:
            grid[0][0] = 1
            queue.append((0,0))
        
        step = 0
        
        while queue:
            step+=1

            next_queue = []

            for i, j in queue:
                
                if i == len(grid) - 1 and j == len(grid[0]) - 1:
                    return step

                if i > 0:
                    if j > 0:
                        if grid[i-1][j-1] == 0:
                            next_queue.append((i-1, j-1))
                            grid[i-1][j-1] = 1
                    if grid[i-1][j] == 0:
                        next_queue.append((i-1, j))
                        grid[i-1][j] = 1
                    if j < len(grid[0]) - 1:
                        if grid[i-1][j+1] == 0:
                            next_queue.append((i-1, j+1))
                            grid[i-1][j+1] = 1
                if j > 0:
                    if grid[i][j-1] == 0:
                        next_queue.append((i, j-1))
                        grid[i][j-1] = 1

                if j < len(grid[0]) - 1:
                    if grid[i][j+1] == 0:
                        next_queue.append((i, j+1))
                        grid[i][j+1] = 1
                                     
                if i < len(grid) - 1:
                    if j > 0:
                        if grid[i+1][j-1] == 0:
                            next_queue.append((i+1, j-1))
                            grid[i+1][j-1] = 1
                    if grid[i+1][j] == 0:
                        next_queue.append((i+1, j))
                        grid[i+1][j] = 1

                    if j < len(grid[0]) - 1:
                        if grid[i+1][j+1] == 0:
                            next_queue.append((i+1, j+1))
                            grid[i+1][j+1] = 1
            
            queue = next_queue

        return -1
```
