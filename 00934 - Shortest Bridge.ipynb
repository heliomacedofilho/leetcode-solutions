```python
class Solution:
    def shortestBridge(self, grid: List[List[int]]) -> int:

        def boundaries(grid, i, j, boundary = set()):
            grid[i][j] = -1
            if i > 0:
                if grid[i-1][j] == 0:
                    boundary.add((i-1, j))
                if grid[i-1][j] == 1:
                    boundary = boundaries(grid, i-1, j, boundary)
            if j > 0:
                if grid[i][j-1] == 0:
                    boundary.add((i, j-1))
                if grid[i][j-1] == 1:
                    boundary = boundaries(grid, i, j-1, boundary)
            if i < len(grid) - 1:
                if grid[i+1][j] == 0:
                    boundary.add((i+1, j))
                if grid[i+1][j] == 1:
                    boundary = boundaries(grid, i+1, j, boundary)
            if j < len(grid[0]) - 1:
                if grid[i][j+1] == 0:
                    boundary.add((i, j+1))
                if grid[i][j+1] == 1:
                    boundary = boundaries(grid, i, j+1, boundary)
            return boundary

        def bridgeSize(grid, current_step):
            step=1
            while (current_step):
                next_step = set()
                for i, j in current_step:
                    grid[i][j] = -1
                    if i > 0:
                        if grid[i-1][j] == 0:
                            next_step.add((i-1, j))
                        if grid[i-1][j] == 1:
                            return step
                    if j > 0:
                        if grid[i][j-1] == 0:
                            next_step.add((i, j-1))
                        if grid[i][j-1] == 1:
                            return step
                    if i < len(grid) - 1:
                        if grid[i+1][j] == 0:
                            next_step.add((i+1, j))
                        if grid[i+1][j] == 1:
                            return step
                    if j < len(grid[0]) - 1:
                        if grid[i][j+1] == 0:
                            next_step.add((i, j+1))
                        if grid[i][j+1] == 1:
                            return step
                current_step = next_step                        
                step+=1
            return step
                
        
        boundary = set()
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 1:
                    boundary = boundaries(grid, i, j)
                    return bridgeSize(grid, boundary)
```
